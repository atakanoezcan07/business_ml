```python
import pandas as pd
import numpy as np
#First read the dataset
path='Business_sales_EDA.csv'
b_sales=pd.read_csv(path,delimiter=';', encoding='utf-8')
#print dataset
b_sales.head(5)
```

```python
b_sales.info()
```

```python
b_sales.isnull().sum()
```

```python
#there is two missing data points for description, but we do not want to know it for every dataset, so I am taking the column into a loop and check isnull sum
columns=b_sales.columns
for column in columns:
    if b_sales[column].isnull().sum()>0:
        b_sales[column]=b_sales[column].fillna('missing value')
b_sales.isnull().sum()
```

```python
#Scaling string in the data set
from sklearn.preprocessing import LabelEncoder
columns=b_sales.columns
for column in columns:
    b_sales[column] = LabelEncoder().fit_transform(b_sales[column])
# if code works we can see all data sets as scaled
b_sales.head(5)
```

```python
#Drop some columns
b_sales=b_sales.drop('Product ID',axis=1)
b_sales=b_sales.drop('url',axis=1)
b_sales.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scale_cols = ['Sales Volume', 'price','name','description']  # only numeric columns here

scaler = MinMaxScaler()
b_sales[scale_cols] = scaler.fit_transform(b_sales[scale_cols])

b_sales.head(20)
```

```python
#First project I will basically train this with linear regression
print(b_sales.columns)
```

```python
input_columns=['Product Position','Promotion','Product Category', 'Seasonal', 'brand','name','description','price','currency','terms','section','season','material','origin']
output_column=['Sales Volume']
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X=b_sales[input_columns]
y=b_sales[output_column]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
reg = LinearRegression().fit(X_train,y_train)
reg.score(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_pred=reg.predict(X_test)
# If y_test is a DataFrame with one column:
y_test_1d = y_test.iloc[:, 0]       # turn into Series
y_pred_1d = np.ravel(y_pred)        # ensure 1D array

plt.figure(figsize=(5, 5))
plt.scatter(y_test_1d, y_pred_1d, alpha=0.5)
plt.xlabel("Actual Sales Volume")
plt.ylabel("Predicted Sales Volume")
plt.title("Actual vs Predicted Sales Volume (Test set)")

# 45-degree line (perfect predictions)
min_val = min(y_test_1d.min(), y_pred_1d.min())
max_val = max(y_test_1d.max(), y_pred_1d.max())
plt.plot([min_val, max_val], [min_val, max_val], linestyle="--")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

y_pred = reg.predict(X_test)
residuals = y_test - y_pred

plt.figure(figsize=(6,4))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(0, linestyle="--")
plt.xlabel("Predicted Sales Volume")
plt.ylabel("Residual (Actual - Predicted)")
plt.title("Residuals vs Predicted")
plt.tight_layout()
plt.show()
```
